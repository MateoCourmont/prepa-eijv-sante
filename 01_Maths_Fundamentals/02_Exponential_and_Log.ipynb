# 2. Exponentielle & Logarithme

## A. La Fonction Exponentielle ($e^x$)
C'est la fonction définie par $f(x) = e^x$ avec $e \approx 2.718$.
Elle est **strictement positive** ($e^x > 0$). Elle ne touche jamais 0.

**Propriétés Vitales (Loi des Puissances) :**
* $e^0 = 1$
* $e^{a+b} = e^a \times e^b$ (L'addition devient multiplication)
* $e^{-a} = \frac{1}{e^a}$

## B. Le Logarithme Népérien ($\ln x$)
C'est simplement l'**Inverse** de l'exponentielle.
* Domaine : Défini uniquement pour $x > 0$.
* $\ln(1) = 0$
* $\ln(e) = 1$

## C. La Relation Clé (L'Annulation)
Comme ce sont des fonctions inverses, elles s'annulent :
$$\ln(e^x) = x$$
$$e^{\ln(x)} = x$$

## D. Propriétés Algébriques (La Boîte à Outils)

Pour manipuler ces fonctions, il faut connaître 3 règles par cœur. Elles transforment les opérations complexes en opérations simples.

| Opération dans le $\ln$ | Devient... | Formule |
| :--- | :--- | :--- |
| **Multiplication** | Addition | $\ln(a \times b) = \ln(a) + \ln(b)$ |
| **Division** | Soustraction | $\ln(\frac{a}{b}) = \ln(a) - \ln(b)$ |
| **Puissance** (CRITIQUE) | Multiplication | $\ln(a^n) = n \cdot \ln(a)$ |
| **Changement de Base** | Transforme $a^x$ en $e^{...}$ | $a^x = e^{x \cdot \ln(a)}$ |

<br>

> **Note :** Attention, $\ln(a + b)$ ne se simplifie pas ! C'est un piège classique.

## E. Résolution d'équations ($x$ en exposant)
Pour trouver $x$ quand il est coincé dans une puissance (ex: $e^{kt} = y$), on applique le logarithme des deux côtés.

**Exemple : Calcul de temps**
On cherche le temps $t$ pour qu'une population double ($e^{0.5t} = 2$).
1. On applique $\ln$ : $\ln(e^{0.5t}) = \ln(2)$
2. On simplifie : $0.5t = \ln(2)$
3. On isole : $t = \frac{\ln(2)}{0.5} \approx 1.38$

## F. Comportement aux limites (Qui gagne ?)

1.  **Aux extrêmes :**
    * En $+\infty$ : $e^x$ tend vers l'infini (Explosion). $\ln(x)$ aussi, mais très lentement.
    * En $-\infty$ : $e^x$ tend vers **0** (Asymptote horizontale).
    * En $0$ (pour $\ln$) : $\ln(x)$ plonge vers $-\infty$ (Asymptote verticale).



2.  **Croissances Comparées (Le Duel) :**
    En $+\infty$, l'exponentielle est **toujours plus forte** que n'importe quel polynôme ($x^2, x^3...$).
    $$\lim_{x \to +\infty} \frac{e^x}{x^n} = +\infty$$
    *Interprétation : Une épidémie ($e^x$) ira toujours plus vite qu'un véhicule ($x^2$).*

## 2.1 Croissance comparée (le Mur exponentiel, $x^2$ vs $e^x$)

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Création des données
x = np.linspace(0, 10, 100)

y_linear = x          # Croissance Linéaire
y_quad = x**2         # Croissance Quadratique
y_expo = np.exp(x)    # Croissance Exponentielle

# 2. Le Match
plt.figure(figsize=(10, 6))

# On trace
plt.plot(x, y_linear, label='Linéaire ($x$)', linestyle='--')
plt.plot(x, y_quad, label='Quadratique ($x^2$)', linewidth=2)
plt.plot(x, y_expo, label='Exponentielle ($e^x$)', linewidth=3, color='red')

# Réglages pour bien voir le e^x
plt.ylim(0, 100)
plt.xlim(0, 6) 
plt.title("L'Exponentielle gagne toujours à la fin")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 2.2 Symétrie miroir (fonctions inverses, $e^x$ vs $ln(x)$)

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Création des données
# Pour l'exponentielle (on peut aller dans les négatifs)
x_exp = np.linspace(-3, 2, 100)
y_exp = np.exp(x_exp)

# Pour le Logarithme (OBLIGATOIRE : x > 0)
# On prend des valeurs qui correspondent aux résultats de l'exponentielle pour la symétrie
x_log = np.linspace(0.1, 7, 100) 
y_log = np.log(x_log)

# La diagonale
x_mirror = np.linspace(-3, 7, 100)
y_mirror = x_mirror

# 2. Le Graphique
plt.figure(figsize=(6, 6))

plt.plot(x_exp, y_exp, label='$e^x$ (Exponentielle)', color='red', linewidth=2)
plt.plot(x_log, y_log, label='$\ln(x)$ (Logarithme)', color='green', linewidth=2)
plt.plot(x_mirror, y_mirror, label='Axe de symétrie $y=x$', color='gray', linestyle='--')

# Réglages esthétiques
plt.title("Symétrie entre $e^x$ et $\ln(x)$")
plt.grid(True, linestyle=':', alpha=0.3)
plt.legend()

# On fixe les limites pour avoir un repère orthonormé
plt.xlim(-3, 5)
plt.ylim(-3, 5)

# Lignes des zéros (axes noirs)
plt.axvline(0, color='black', linewidth=1)
plt.axhline(0, color='black', linewidth=1)

plt.show()
```

### Exercice 1 : Le Crash du Serveur (Loi de Newton)
Suite à une panne de ventilateur, le processeur d'un serveur de calcul a atteint une température critique de **80°C**.
Par sécurité, le système s'est arrêté d'urgence.

Le serveur refroidit maintenant passivement dans une salle climatisée maintenue à **20°C**.
La physique nous dit que la vitesse de refroidissement suit une **décroissance exponentielle** (Loi de Newton).

### Les Données
* **Température Initiale ($T_{départ}$)** : $80^\circ C$
* **Température Ambiante ($T_{env}$)** : $20^\circ C$
* **Coefficient de refroidissement ($k$)** : $0.15$ (propre au matériau du dissipateur thermique).
* **Seuil de redémarrage ($T_{sécurité}$)** : Le serveur ne peut être rallumé que lorsqu'il descend sous **30°C**.

### La Modélisation
La température $T(t)$ au temps $t$ (en minutes) est donnée par la formule :
$$T(t) = T_{env} + (T_{départ} - T_{env}) \cdot e^{-k \cdot t}$$

### Objectifs
1.  **Calculer** précisément (via la résolution d'équation logarithmique) au bout de combien de minutes le serveur pourra être rallumé.
2.  **Visualiser** la courbe de température pour les 30 prochaines minutes.

#### Calcul manuel :

$$
\begin{aligned}
& T = T_{env} + (T_{depart} - T_{env}) \cdot e^{-k \cdot t} \\
& \Leftrightarrow T - T_{env} = (T_{depart} - T_{env}) \cdot e^{-k \cdot t} \\
& \Leftrightarrow 30 - 20 = (80 - 20) \cdot e^{-0,15 t} \\
& \Leftrightarrow 10 = 60 \cdot e^{-0,15 t} \\
& \Leftrightarrow \frac{10}{60} = e^{-0,15 t} \\
& \Leftrightarrow \ln\left(\frac{1}{6}\right) = -0,15 t \\
& \Leftrightarrow t = \frac{\ln(1/6)}{-0,15} \\
& \text{Application numérique :} \\
& t \approx \frac{-1,79}{-0,15} \\
& t \approx 11,93 \text{ min}
\end{aligned}
$$


#### Visualisation de la courbe

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Paramètres physiques
T_env = 20       
T_depart = 80    
k = 0.15       
T_securite = 30 

# 2. Temps (de 0 à 30 minutes)
t = np.linspace(0, 30, 100)

# 3. La Formule
T_t = T_env + (T_depart - T_env) * np.exp(-k * t)

# 4. Calcul du temps exact pour atteindre la température de sécurité
t_solution = np.log((T_securite - T_env) / (T_depart - T_env)) / -k

# 5. Graphique
plt.figure(figsize=(10, 6))

# Température (T_t)
plt.plot(t, T_t, label='Température du Serveur', color='blue', linewidth=2)

# Ligne de sécurité (T_securite)
plt.axhline(T_securite, color='red', linestyle='--', label='Seuil de Sécurité')

plt.title("Refroidissement du Serveur")
plt.xlabel("Temps (minutes)")
plt.ylabel("Température (°C)")
plt.grid(True, alpha=0.3)
plt.legend()

plt.scatter([t_solution], [T_securite], color='red', zorder=5)

plt.show()
```

### Exercice 2 : Le Réveil du Patient (Pharmacocinétique)

Tu es anesthésiste. Tu as injecté une dose de **Propofol** (anesthésiant puissant) à un patient pour une opération longue.
Le corps du patient élimine le produit naturellement via le foie et les reins (filtrage).

Si la concentration dans le sang descend en dessous d'un certain seuil, **le patient se réveille** en pleine opération. Tu dois calculer précisément quand réinjecter une dose.

### Les Données
* **Concentration Initiale ($C_0$)** : $120 \text{ mg/L}$ (Bolus de départ).
* **Vitesse d'élimination ($k$)** : $0.35$ (par heure).
* **Seuil de réveil ($C_{min}$)** : Le patient se réveille si la concentration passe sous **$15 \text{ mg/L}$**.

### La Modélisation
La concentration $C(t)$ au temps $t$ (en heures) suit une décroissance exponentielle simple :
$$C(t) = C_0 \cdot e^{-k \cdot t}$$

### Objectifs
1.  **Maths & Code :** Résous l'équation pour trouver l'heure exacte du réveil ($t_{réveil}$).
2.  **Code :** Trace la courbe de concentration du médicament sur **8 heures**.
3.  **Code :** Ajoute une ligne rouge horizontale à **15 mg/L** (Zone de danger).

#### Calcul manuel :

$$
\begin{aligned}
&C = C_0 \cdot e^{-k \cdot t} \\
& \Leftrightarrow 15 = 120 \cdot e^{-0,35t} \\
& \Leftrightarrow \frac{15}{120} = e^{-0,35t} \\
& \Leftrightarrow \ln(\frac{15}{120}) = -0,35t \\
& \Leftrightarrow t = \frac{\ln(15/120)}{-0,35} \\
& \text{Application numérique :} \\
& t \approx \frac{-2,079}{-0,35} \\
& t \approx 5,94
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Données
C0 = 120        # Dose départ
k = 0.35        # Vitesse élimination
C_min = 15      # Seuil réveil

# 2. Temps (0 à 8 heures)
t = np.linspace(0, 8, 100)

# 3. Formule
C_t = C0 * np.exp(-k * t)

# 4. Calcul exact du réveil
t_reveil = np.log(C_min / C0) / -k

# 5. Graphique
plt.figure(figsize=(10, 6))

plt.plot(t, C_t, label='Concentration Propofol', color='purple', linewidth=2)
plt.axhline(C_min, color='red', linestyle='--', label='Seuil de Réveil (15 mg/L)')

plt.title(f"Pharmacocinétique : Réveil au bout de t = {t_reveil:.2f} heures")
plt.xlabel("Temps (heures)")
plt.ylabel("Concentration (mg/L)")
plt.grid(True, alpha=0.3)
plt.legend()

plt.scatter([t_reveil], [C_min], color='red', zorder=5)

plt.show()
```

### Exercice 3 : L'Invasion de la Boîte de Pétri

Tu cultives une souche de bactéries (E. Coli) dans une boîte de Pétri pour tester un antibiotique.
Au début de l'expérience, la population est faible, mais ces bactéries se divisent très vite.

Tu dois impérativement injecter l'antibiotique avant que la population n'atteigne le seuil critique de **saturation**, sinon l'expérience est fichue (le milieu de culture sera épuisé).

### Les Données
* **Population Initiale ($N_0$)** : 100 bactéries.
* **Taux de croissance ($k$)** : $0.7$ par heure.
* **Seuil critique ($N_{max}$)** : 50 000 bactéries.

### La Modélisation
La population $N(t)$ au temps $t$ (en heures) suit une croissance exponentielle (Loi de Malthus) :
$$N(t) = N_0 \cdot e^{k \cdot t}$$


### Objectifs
1.  **Maths & Code :** Calcule précisément à quelle heure tu dois intervenir au plus tard.
2.  **Code :** Trace la courbe de prolifération sur **15 heures**.
3.  **Code :** Ajoute la ligne rouge du seuil critique (50 000).

#### Calcul manuel 

$$
\begin{aligned}
& N = N_{0} \cdot e^{k \cdot t} \\
& \Leftrightarrow 50000 = 100 \cdot e^{0,7t} \\
& \Leftrightarrow \frac{50000}{100} = e^{0,7t} \\
& \Leftrightarrow 500 = e^{0,7t} \\
& \Leftrightarrow \ln(500) = 0,7t \\
& \text{Application numérique :} \\
& t \approx \frac{6,215}{0,7} \\
& t \approx 8,88
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Données
N0 = 100        # Population départ
k = 0.7         # Taux de croissance
N_max = 50000   # Seuil critique

# 2. Temps (0 à 12 heures)
t = np.linspace(0, 12, 100)

# 3. Formule
N_t = N0 * np.exp(k * t)

# 4. Calcul précis
t_limite = np.log(N_max / N0) / k

# 5. Graphique
plt.figure(figsize=(10, 6))

plt.plot(t, N_t, label='Bactéries E. Coli', color='green', linewidth=2)
plt.axhline(N_max, color='red', linestyle='--', label='Saturation (50k)')

plt.scatter([t_limite], [N_max], color='red', zorder=5, s=30, label='Point de rupture')

plt.title(f"Invasion Bactérienne : Saturation à t = {t_limite:.2f} h")
plt.xlabel("Temps (heures)")
plt.ylabel("Population (nombre de bactéries)")
plt.grid(True, alpha=0.3)
plt.legend()

plt.show()
```
