# 1. Fonctions du Second Degré

## A. Définitions
La forme standard (développée) d'une fonction de degré 2 est :
$$f(x) = ax^2 + bx + c$$

Elle peut aussi s'écrire sous **Forme Canonique** pour visualiser immédiatement le sommet : 
$$f(x) = a(x - \alpha)^2 + \beta$$
Où le sommet $S$ se trouve aux coordonnées $(\alpha, \beta)$.

## B. Le Discriminant et les Racines
Pour trouver les racines (solutions pour $f(x)=0$), il faut d'abord calculer le discriminant $\Delta$ :
$$\Delta = b^2 - 4ac$$

* **Cas 1 : $\Delta > 0$** (2 solutions réelles distinctes)
    La parabole coupe l'axe des abscisses en deux points :
    $$x_1 = \frac{-b - \sqrt{\Delta}}{2a} \quad \text{et} \quad x_2 = \frac{-b + \sqrt{\Delta}}{2a}$$

* **Cas 2 : $\Delta = 0$** (1 solution réelle unique)
    Le sommet de la parabole touche juste l'axe des abscisses :
    $$x_0 = \frac{-b}{2a}$$

* **Cas 3 : $\Delta < 0$** (Pas de solution réelle)
    La courbe ne touche jamais l'axe des abscisses.
    *(Nécessite les nombres complexes pour être résolu)*.

## C. Étude du Signe (Inégalités)

L'objectif est de savoir quand la fonction est positive (au-dessus de l'axe des x) ou négative (en dessous).

### La Règle d'Or (Algorithme)
> **"Le polynôme est du SIGNE DE $a$ à l'extérieur des racines."**

---

### Cas 1 : $\Delta > 0$ (Deux racines $x_1$ et $x_2$)
La courbe traverse l'axe deux fois.

| $x$ | $-\infty$ | | $x_1$ | | $x_2$ | | $+\infty$ |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| **Signe de $f(x)$** | | **Signe de $a$** | $0$ | Signe opposé de $a$ | $0$ | **Signe de $a$** | |

* **Si $a > 0$ (La parabole sourit) :** Elle est Positive, puis Négative, puis Positive.
* **Si $a < 0$ (La parabole pleure) :** Elle est Négative, puis Positive, puis Négative.

---

### Cas 2 : $\Delta = 0$ (Une racine unique $x_0$)
La courbe touche juste l'axe (rebond).

| $x$ | $-\infty$ | | $x_0$ | | $+\infty$ |
| :---: | :---: | :---: | :---: | :---: | :---: |
| **Signe de $f(x)$** | | **Signe de $a$** | $0$ | **Signe de $a$** | |

* Le signe ne change jamais, sauf au point de contact où c'est nul.

---

### Cas 3 : $\Delta < 0$ (Pas de racine)
La courbe ne touche jamais l'axe.

| $x$ | $-\infty$ | | $+\infty$ |
| :---: | :---: | :---: | :---: |
| **Signe de $f(x)$** | | **Toujours du Signe de $a$** | |

* Si $a > 0$, tout est positif.
* Si $a < 0$, tout est négatif.

## D. Variations & Extremum (Minimum / Maximum)

Le signe de $a$ détermine l'orientation de la parabole et la nature de son sommet $S(\alpha, \beta)$.

* **Cas 1 : $a > 0$ (Convexe - La parabole "sourit")**
    * La fonction est décroissante puis croissante ($\searrow \nearrow$).
    * Le sommet est un **MINIMUM Global**.
    * La valeur minimale atteinte est **$\beta$**.

* **Cas 2 : $a < 0$ (Concave - La parabole "pleure")**
    * La fonction est croissante puis décroissante ($\nearrow \searrow$).
    * Le sommet est un **MAXIMUM Global**.
    * La valeur maximale atteinte est **$\beta$**.

> **Note Ingénieur (Optimisation) :**
> Si cette fonction modélise un **Coût**, on cherche un cas où $a > 0$ pour minimiser les dépenses.
> Si elle modélise un **Gain**, on cherche un cas où $a < 0$ pour maximiser les profits.

## 1.1. Calcul de fonction du second degré par code Python

```python
# Installation des bibliothèques nécessaires
%pip install numpy matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_quadratic(a, b, c): 
    """
    Calcule le discriminant, les racines et le sommet de la parabole d'une fonction polynôme de degré 2.
    """
    # Calcul du discriminant
    D = b**2 - 4*a*c

    print(f"Function: {a}x² + {b}x + {c}")
    print(f"Discriminant (Delta): {D}")

    # Calcul des racines
    if D > 0:
        x1 = (-b - np.sqrt(D)) / (2*a)
        x2 = (-b + np.sqrt(D)) / (2*a)
        print(f"Deux solutions: x1 = {x1:.2f}, x2 = {x2:.2f}")
    elif D == 0:
        x0 = -b / (2*a)
        print(f"Une solution: x0 = {x0:.2f}")
    else:
        print("Pas de solution réelle (Delta < 0)")

    # Calcul du sommet de la parabole
    alpha = -b / (2*a)
    beta = -D / (4*a)
    print(f"Sommet : ({alpha:.2f}, {beta:.2f})")

    return alpha, beta

# Test avec une fonction simple : 2x² - 8x + 6
# On attend un sommet en (2, -2) et des racines en 1 et 3
alpha_found, beta_found = analyze_quadratic(2, -8, 6)
```

## 1.2. Visualisation de la courbe via matplotlib

```python
# Paramètres de la fonction
a, b, c = 2, -8, 6

# 1. Préparer les données X
# On centre le graphique autour du sommet (alpha +/- 5 unités)
x = np.linspace(alpha_found - 5, alpha_found + 5, 100)
y = a*x**2 + b*x + c

# 2. Création du Graphique
plt.figure(figsize=(8, 5))

# La courbe
plt.plot(x, y, label=f'f(x) = {a}x² + {b}x + {c}', color='blue')

# L'axe des abscisses (zéro)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Le Sommet
plt.scatter([alpha_found], [beta_found], color='red', zorder=5)
plt.text(alpha_found, beta_found + 1, f"({alpha_found:.2f}, {beta_found:.2f})", color='red', ha='center', va='bottom', fontsize=12)

# Décoration
plt.title("Visualisation de la Parabole")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True, linestyle='-', alpha=0.3)
plt.legend()

plt.show()
```
